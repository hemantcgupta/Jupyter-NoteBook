# Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'x': [1, 2, 3, 4, 5],
        'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)

# Create a line plot
plt.plot(df['x'], df['y'])
plt.scatter(df['x'], df['y'])

# Set the title and axis labels
plt.title("Line Plot Example")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'x': [1, 2, 3, 4, 5],
        'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)

# Create a line plot using Seaborn
sns.lineplot(x=df['x'], y=df['y'])
sns.scatterplot(x=df['x'], y=df['y'])

# Set the title and axis labels
plt.title("Line Plot Example")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the plot
plt.show()
```

# MultiLine Plot

```python
import matplotlib.pyplot as plt

# Create some data for three lines
data = {'x':[1, 2, 3, 4, 5],
        'y1':[2, 4, 6, 8, 10],
        'y2':[1, 3, 5, 7, 9],
        'y3':[3, 6, 9, 12, 15]}
df = pd.DataFrame(data)
# Create a figure and axis object
fig, ax = plt.subplots()

# Create multiline plot using the plot() function
ax.plot(df['x'], df['y1'], label='Line 1')
ax.plot(df['x'], df['y2'], label='Line 2')
ax.plot(df['x'], df['y3'], label='Line 3')

# Set the title and axis labels
ax.set_title("Multiline Plot Example")
ax.set_xlabel("X-axis Label")
ax.set_ylabel("Y-axis Label")

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create some data for three lines
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [3, 6, 9, 12, 15]

# Combine data into a single DataFrame
df = pd.DataFrame({'x': x, 'Line 1': y1, 'Line 2': y2, 'Line 3': y3})

# Melt the data to create a long-form DataFrame
df_long = df.melt(id_vars='x', var_name='Line', value_name='Y')

# Create multiline plot using the lineplot() function
sns.lineplot(x='x', y='Y', hue='Line', data=df_long)

# Set the title and axis labels
plt.title("Multiline Plot Example")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the plot
plt.show()
```

# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure and two subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Create a line plot on the first subplot
axs[0].plot(x, y1, label='Sin(x)')
axs[0].plot(x, y2, label='Cos(x)')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].set_title('Sin and Cos Functions')
axs[0].legend()

# Create a scatter plot on the second subplot
axs[1].scatter(x, y3, marker='.', color='red')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')
axs[1].set_title('Tan Function')

# Adjust spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")
print(tips.head())
# Create a figure and two subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Create a boxplot on the first subplot
sns.boxplot(x="day", y="total_bill", data=tips, ax=axs[0])
axs[0].set_xlabel('Day')
axs[0].set_ylabel('Total Bill')
axs[0].set_title('Total Bill by Day')

# Create a violinplot on the second subplot
sns.violinplot(x="day", y="tip", data=tips, ax=axs[1])
axs[1].set_xlabel('Day')
axs[1].set_ylabel('Tip')
axs[1].set_title('Tip by Day')

# Adjust spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Calculate the total tips by gender
tips_by_gender = tips.groupby('sex').sum()['tip']

# define Seaborn color palette to use
palette_color = sns.color_palette('bright')

# Create a pie chart
plt.pie(tips_by_gender, labels=tips_by_gender.index, colors=palette_color, autopct='%1.1f%%')
plt.title('Pie Chart of Total Tips by Gender')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Calculate the total tips by gender
tips_by_gender = tips.groupby('sex').sum()['tip']

# define Seaborn color palette to use
palette_color = sns.color_palette('dark')

# declaring exploding pie
explode = [0, 0.1]

# Create a pie chart
plt.pie(tips_by_gender, labels=tips_by_gender.index, colors=palette_color, explode=explode, autopct='%1.1f%%')
plt.title('Pie Chart of Total Tips by Gender')

# Show the plot
plt.show()
```

# Joint Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)

# Create a joint plot of the data
plt.hist2d(x, y, bins=30, cmap='Reds')

# Add axis labels
plt.xlabel('X')
plt.ylabel('Y')

# Add a colorbar
plt.colorbar()

# Add a title
plt.title('Joint Plot of X and Y')

# Show the plot
plt.show()
```

```python
import seaborn as sns

# Load some example data
tips = sns.load_dataset("tips")

# Create a joint plot of the data
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter")

# Add a title
plt.subplots_adjust(top=0.9)
plt.suptitle('Joint Plot of Total Bill and Tip')

# Show the plot
plt.show()
```

# Swarm Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a random dataset
data = np.random.normal(size=(10,3))

# Create a figure and axis object
fig, ax = plt.subplots()

# Create the swarm plot
ax.scatter(data[:,0], np.zeros(data.shape[0]), marker='o', label='Column 1')
ax.scatter(data[:,1], np.ones(data.shape[0]), marker='s', label='Column 2')
ax.scatter(data[:,2], 2*np.ones(data.shape[0]), marker='^', label='Column 3')

# Add labels and legend
ax.set_xlabel('Values')
ax.set_yticks([0,1,2])
ax.set_yticklabels(['Column 1', 'Column 2', 'Column 3'])
ax.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import numpy as np

# Create a random dataset
data = np.random.normal(size=(10,3))

# Create the swarm plot using Seaborn
sns.swarmplot(data=data, size=8)

# Show the plot
plt.show()
```

# Pair Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Load the iris dataset
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

# Create a scatter plot matrix
scatter_matrix(iris, alpha=0.2, figsize=(10, 10), diagonal='hist')

# Add a title
plt.suptitle('Pair Plot of Iris Dataset', size=16)

# Show the plot
plt.show()
```

# Heatmap

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a random dataset
data = np.random.rand(10, 10)

# Create a heatmap with numbers
fig, ax = plt.subplots()
im = ax.imshow(data, cmap='coolwarm', interpolation='nearest')

# Add numbers to the heatmap
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        text = ax.text(j, i, data[i, j].round(2), ha="center", va="center", color="w")

# Add a colorbar
plt.colorbar(im)

# Show the plot
plt.show()
```
